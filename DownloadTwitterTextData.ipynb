```python
def collect_company_tweets(api, q, lang='en', save_path=None, until=None):
    tweets = tweepy.Cursor(api.search, q=q, lang=lang, until=until)
    
    collected_texts = pd.Series()
    for status in tweets.items():
        collected_texts[status.created_at] = status.text
        if status.created_at.strftime("%Y-%m-%d") == '2020-11-25': break
        
    collected_texts.to_csv(save_path+'/'+q+".csv")
    
    return collected_texts

ex=collect_company_tweets(api, 'J.P. Morgan', 
                          save_path="F:/Graduate/Lecture Notes and Homework/NLP",
                         until="2020-12-09")    
ex
```
